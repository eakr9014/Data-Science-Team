```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import string
```

Read in test and training data

```python
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
train_df.head()
```

Let's try to visualize the most common keywords

```python
plt.figure(figsize=(9,6))
sns.countplot(y=train_df.keyword, order = train_df.keyword.value_counts().iloc[:15].index)
plt.title('Top 15 keywords')
plt.show()
```

```python
kw_d = train_df[train_df.target==1].keyword.value_counts().head(10)
kw_nd = train_df[train_df.target==0].keyword.value_counts().head(10)

plt.figure(figsize=(13,5))
plt.subplot(121)
sns.barplot(kw_d, kw_d.index, color='m')
plt.title('Top keywords for disaster tweets')
plt.subplot(122)
sns.barplot(kw_nd, kw_nd.index, color='c')
plt.title('Top keywords for non-disaster tweets')
plt.show()
```

```python
import re
from nltk.corpus import stopwords

my_stopwords = stopwords.words('english')

def text_process(text):
    text = re.sub(r'https?://\S+', '', text) # Remove links
    text = re.sub(r'\n',' ', text) # Remove line breaks
    text = re.sub('\s+', ' ', text).strip() # Remove extra spaces
    text = [char for char in text if char not in string.punctuation] # remove punctuation
    text = ''.join(text)
    text = [word for word in text.split() if word not in my_stopwords] # remove common words
    text = ''.join(text)
    return text

test_str = train_df.loc[417, 'text']
print(test_str)
print(text_process(test_str))
```
